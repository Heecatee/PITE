```python
import numpy as npy
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sb
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
```

```python
boston_dataset = load_boston()
data_frame = pd.DataFrame(data = boston_dataset.data, columns = boston_dataset.feature_names)
```

```python
data_frame.head()
```

```python
nox_pl = data_frame['NOX']
age_pl = data_frame['AGE']
```

```python
figure, ax = plot.subplots(1,2)
sb.distplot(nox_pl, ax=ax[0], hist=True)
sb.distplot(age_pl, ax=ax[1], hist=True)
```

```python
price_ds = boston_dataset.target
```

```python
def plotter(x, y, linear_model):
    linear_model.fit(x.reshape((-1,1)), y)
    sb.scatterplot(x, y)
    
    # r-squared to tell how data fit into model 
    # https://blog.minitab.com/blog/adventures-in-statistics-2/regression-analysis-how-do-i-interpret-r-squared-and-assess-the-goodness-of-fit
    r_squared = linear_model.score(x.reshape((-1,1)), y)
    
    # prediction
    new_y = linear_model.predict(x.reshape((-1,1)))
    sb.lineplot(x, new_y)
    print('R-squared: ', r_squared)
```

## 1. Two plots of picked columns

**1. NOX**

```python
plotter(nox_pl.values, price_ds, LinearRegression())
```

**2. AGE**

```python
plotter(age_pl.values, price_ds, LinearRegression())
```

**Conclusion:** Based on R-squared parameter, both of those models are not well fitted with the data I had.
